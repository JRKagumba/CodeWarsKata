```python
import numpy as np


clues = ( 2, 2, 1, 3,  
  2, 2, 3, 1,  
  1, 2, 2, 3,  
  3, 2, 1, 4 )

Dict={}

units=[]

chars = 'ABCD'
nums = '1234'

lst = [1,2,3,4]


arr = [[chars[j]+nums[i] for i in range(4)] for j in range(4)]
board = [[0 for i in range(4)] for j in range(4)]

print(np.matrix(arr))

for i in chars:
    for j in nums:
        Dict[i+j]={}
        Dict[i+j]['peers']=[]
        Dict[i+j]['lst']=lst
        
#Unit Designation
#Rows
for i in range(4):
    units.append(arr[i])
#Cols
for i in range(4):
    units.append([arr[j][i] for j in range(4)])
    
#Creating a list of peers    
for i in Dict.keys():
    for unit in units:
        if i in unit:
            for j in unit:
                if i!=j: 
                    Dict[i]['peers'].append(j)
          
seq = []

for i in range(4):
    seq.append([[arr[j][i],[1,2,3,4]] for j in range(4)])
for i in range(4):
    seq.append([[arr[i][j],[1,2,3,4]] for j in reversed(range(4))])
for i in reversed(range(4)):
    seq.append([[arr[j][i],[1,2,3,4]] for j in reversed(range(4))])
for i in reversed(range(4)):
    seq.append([[arr[i][j],[1,2,3,4]] for j in range(4)])
  


Clues_Dict = {}

for i in range(len(clues)):
    Clues_Dict[i]={}
    Clues_Dict[i]['Value']=clues[i]
    Clues_Dict[i]['List'] = seq[i]
    

    
for i in Clues_Dict:
    
    if Clues_Dict[i]['Value'] == 4:
        for j, jv in enumerate(Clues_Dict[i]['List']):    
            Clues_Dict[i]['List'][j][1]=[j+1]
            
Clues_Dict   
```
